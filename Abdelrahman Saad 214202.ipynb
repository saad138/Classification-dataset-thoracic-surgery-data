```python
from sklearn.metrics import roc_curve, roc_auc_score

def best_threshold(y_test, y_pred):
# get the fpr, tpr, and thresholds arrays from the roc_curve http://localhost:8888/notebooks/Abalone/Abalone.ipynb#2--Best-threshold,-TPR,-FPR,----AUC,-F1-and----ACCfunction
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
#Evalute auc from roc_auc_score    
    auc = roc_auc_score(y_test, y_pred)
   
    best_thresh = None
    best_tpr = None
    best_fpr = None
    
# compute the Euclidean distance from the top-left corner for each point on the ROC curve
    D = np.sqrt((1 - tpr) ** 2 + fpr ** 2)
    ix = np.argmin(D)
# find the optimal threshold, tpr, and fpr values 
    best_thresh = thresholds[ix]
    best_tpr = tpr[ix]
    best_fpr = fpr[ix]

    print('The Euclidean distance method minimizes the distance from the top-left corner of the ROC plot, which corresponds to the ideal point of TPR = 1 and FPR = 0.')


   
    return best_thresh, best_tpr, best_fpr
      
```

```python
from sklearn.metrics import f1_score,accuracy_score

def compare_thresholds(clfs): 
    df = pd.DataFrame(columns=['Classifier', 'Threshold', 'TPR', 'FPR', 'AUC', 'F1'])
# fit the classifier on the train set
    for clf in clfs:
        clf.fit(X_train, y_train) 
# get the probability estimates for the test set        
        probas_ = clf.predict_proba(X_test)
    
#get the outputs from  best_threshold function      
        best_thresh, best_tpr, best_fpr = best_threshold(y_test, probas_[:, 1])
# Evalute the f1 score from the f1_score function using 0.5 as the default threshold  
        f1 = f1_score(y_test, probas_[:, 1] > best_thresh)
# Evalute the auc score from the roc_auc_score function
        auc = roc_auc_score(y_test, probas_[:, 1]> best_thresh)
# Evalute accuracy from accuracy_score function     
        acc = accuracy_score(y_test,probas_[:,1]>best_thresh)
         
        df = df.append({'Classifier': clf.__class__.__name__, 'Threshold': best_thresh, 'TPR': best_tpr, 'FPR': best_fpr, 'AUC': auc, 'F1': f1,'ACC': acc}, ignore_index=True)
     
    return df
```

```python
# Compare between the 6 models
classifiers = [LogisticRegression(), DecisionTreeClassifier(), SVC(kernel='linear', probability=True, random_state=42), RandomForestClassifier(),AdaBoostClassifier(RFC), BaggingClassifier(
    base_estimator = DT, n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1, random_state=42),VotingClassifier(estimators=[('svc',
                              SVC(kernel='linear', probability=True,
                                  random_state=42)),
                             ('random_forest',
                              RandomForestClassifier(max_leaf_nodes=7,
                                                     n_estimators=191,
                                                     n_jobs=-1,
                                                     random_state=42)),
                             ('logistic_regression',
                              LogisticRegression(random_state=42)),
                             ('decision tree',
                              DecisionTreeClassifier(max_depth=2,
                                                     random_state=42))],
                 voting='soft')]
compare_thresholds(classifiers)
```

```python
#SMOTE METHOD
#SMOTE METHOD
#SMOTE METHOD

from sklearn.metrics import roc_curve, auc

# List of classifier models to evaluate
clfs = [RandomForestClassifier(random_state=42)]
# Dictionary to store ROC curves
roc_curves = {}

# Evaluate each classifier
for clf in clfs:
    name = clf.__class__.__name__
    clf.fit(x_trainii, y_trainii)
    y_probs = clf.predict_proba(x_testii)[:, 1]
    fpr, tpr, _ = roc_curve(y_testii, y_probs)
    roc_auc = auc(fpr, tpr)
    roc_curves[name] = (fpr, tpr, roc_auc)

# Plot ROC curves
plt.figure(figsize=(8, 6))
for name, (fpr, tpr, auc_score) in roc_curves.items():
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc_score:0.2f})')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.title('ROC Curve for SMOTE')
plt.savefig('roc_comparison.png')
```
